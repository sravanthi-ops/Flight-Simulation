```python
# import random library to generate random numbers  
import random

# Initialize variables , you have 5 planes through out the program and each plane has 100 gallons
#you gave a list for fuels , write it in the file 
planes = 5
fuel_levels = [100,100,100,100,100] # 5 planes with each fuel level of 100
log_file = open("FlightSimulator.txt", "w")

# You are using log function to write the messages in the file 

def log(message):
    # Log messages to the FlightSimulator.txt file  
    log_file.write(message + '\n')

#You are defining display menu function to display the first menu 

def display_menu():
    print(f"There are {planes} planes in the Sky.")
    print("a. Land the Plane")
    print("b. Reduce Fuel")
    print("c. Quit Program")

# You are defining display fuel levels function to display the fuel levels
#you are using for loop for range of planes (5), you are using i+1 to consider indices
#you are selecting a plane and display its fuel level

def display_fuel_levels():
    for i in range(planes):
        print(f"Plane {i+1} - Fuel Level: {fuel_levels[i]} Gallons")

# You are defining display land plane function to land the plane 
#you are using  pop function to remover a plane
# write it in the file

def land_plane(plane_number):
    fuel = fuel_levels.pop(plane_number - 1)
    log(f"Plane {plane_number} has landed. Fuel remaining: {fuel} Gallons.")
    global planes
    planes -= 1

# You are defining reduce fuel function to reduce the fuel  
# you are using count to represent no of planes
# it will select no of planes randomly , then function enters the loop for reducing the fuel 
#if fuel is =>15 , consider fuel for current plane reduce the fuel else give error message 
#display the fuel level after reducing 
# write it in the file

def reduce_fuel(count):
    planes_to_reduce = random.sample(range(planes), count)
    for plane_number in planes_to_reduce:
        fuel = fuel_levels[plane_number]
        if fuel >= 15:
            fuel_levels[plane_number] -= 15
            log(f"Plane {plane_number + 1} Fuel Level reduced by 15 Gallons. New fuel level: {fuel_levels[plane_number]} Gallons.")
        else:
            log(f"Plane {plane_number + 1} is out of fuel and cannot be reduced further.")
    display_fuel_levels()

# we are entering the main function , while planes are>0 , display the initial menu and ask user for the choice
#Use if , elif , elif for a , b , c 

# Main program loop
while planes > 0:
    display_menu()
    choice = input("Enter your Choice: ")
    log(f"User input: {choice}")

# a call display to display the fuel levels , ask user for input for landing the plane
#write it in file , if 1<=plan to land <= planes available land the plane else invalid 

    if choice == 'a':
        display_fuel_levels()
        plane_to_land = int(input("Which Plane Should Land: "))
        log(f"User input: {plane_to_land}")
        if 1 <= plane_to_land <= planes:
            land_plane(plane_to_land)
        else:
            print("Invalid plane number. Please select a valid plane.")

# b # a call display to display the fuel levels , ask user for input for reducing the fuel
#write it in file , if reduction count  > planes reduce else invalid  


    elif choice == 'b':
        display_fuel_levels()
        fuel_reduction_count = int(input("How many Planes Should have their Fuel Level Reduced: "))
        log(f"User input: {fuel_reduction_count}")
        if fuel_reduction_count > planes:
            print("Invalid count. There are fewer planes in the sky.")
        else:
            reduce_fuel(fuel_reduction_count)

# c End the program , write it in the file

    elif choice == 'c':
        print("Flight Simulator program has ended.")
        break

log_file.close()

```
